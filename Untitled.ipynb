```python
import re
df3=df2[df2.total_sqft.apply(lambda x: re.search("\d - \d",x)==None)]
df_with_total_sqft_interval=df2[df2.total_sqft.apply(lambda x: re.search("\d - \d",x)!=None)]
df_with_total_sqft_interval.shape,df3.shape
```

```python
def func(x):
    # Split the range and convert values to floats
    numbers = x.split(" - ")
    numbers = np.array(numbers, dtype='float')
    return np.mean(numbers)

df_with_total_sqft_interval['total_sqft'] = df_with_total_sqft_interval['total_sqft'].apply(lambda x: func(x))
df_with_total_sqft_interval.head()
```

```python
df4 = pd.concat([df_with_total_sqft_interval, df3], axis=0)
df4.reset_index(drop=True, inplace=True)
df4.head()
```

```python
def convert_to_sqft(x):
    if isinstance(x, str):
        # Define conversion factors for different units
        unit_conversions = {
            'Sq. Meter': 10.7639,  # 1 sq. meter = 10.7639 sq. feet
            'Sq. Yards': 9,        # 1 sq. yard = 9 sq. feet
            'Cents': 43560,       # 1 cent = 43560 sq. feet
        }
        
        # Extract numeric part
        numeric_part = ''.join(filter(str.isdigit, x))
        
        # Extract unit
        unit = ''.join(filter(str.isalpha, x))
        
        # Convert to square feet
        if unit in unit_conversions:
            return float(numeric_part) * unit_conversions[unit]
        else:
            return float(numeric_part)
    else:
        # If it's already a numeric value, return it as is
        return x

# Apply the conversion function to the 'total_sqft' column
df4['total_sqft'] = df4['total_sqft'].apply(convert_to_sqft)
df4.describe()
```

<b>Remove outliers total_sqft</b>

```python
Q1 = df4['total_sqft'].quantile(0.25)
Q3 = df4['total_sqft'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.96 * IQR
upper_bound = Q3 + 3 * IQR

df4=df4[(df4['total_sqft'] > lower_bound) & (df4['total_sqft'] < upper_bound)]
# set(df4[(df4['total_sqft'] < lower_bound) | (df4['total_sqft'] > upper_bound)].total_sqft)
```

<b>normally square ft per bedroom is 300 (i.e. 2 bhk apartment is minimum 600 sqft. If you have for example 400 sqft apartment with 2 bhk than that seems suspicious and can be removed as an outlier. We will remove such outliers by keeping our minimum thresold per bhk to be 300 sqft</b>
